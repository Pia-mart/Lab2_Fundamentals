```python
SURNAME = "MARTINEZ"
SEED_NUM = 2  # Last digit of TUPM-25-0602

# --- MISSION 1: REDACTION PROTOCOL ---
redacted_name = ""
vowels = "AEIOU"

print("--- Redaction Execution Log ---")
for char in SURNAME.upper():
    if char in vowels:
        redacted_name += str(SEED_NUM)
        print(f"{char} (Vowel) -> {SEED_NUM}")
    else:
        redacted_name += char
        print(f"{char} (Consonant) -> {char}")

print(f"\nFinal Redacted Output: {redacted_name}")
print("-" * 30)
```

```python
first_letter = SURNAME[0].upper()
ascii_val = ord(first_letter)

# 2. Total loop cycles from Mission 1
loops = len(SURNAME)

# 3. Calculate verification hash
# Formula: (ASCII * SEED_NUM) - loops
verification_hash = (ascii_val * SEED_NUM) - loops

print(f"First Letter: {first_letter} (ASCII: {ascii_val})")
print(f"Total Loops: {loops}")
print(f"Forensic Signature Hash: {verification_hash}")
```
